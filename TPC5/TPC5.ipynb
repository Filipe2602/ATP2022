```python

import pandas as pd

idade= []
sexo= []
tensao= []
colesterol=[]
batimento=[]
temdoenca=[]
paciente=[]





def lerlista():
    file= open("myheart.csv","r")
    file.readline()
    y=0
    for linha in file:
        lista_nova=linha.split(",")
        idade.append(lista_nova[0])
        sexo.append(lista_nova[1])
        tensao.append(lista_nova[2])
        colesterol.append(lista_nova[3])
        batimento.append(lista_nova[4])
        temdoenca.append(lista_nova[5].replace("\n",""))

    
    N=len(idade)
    while N>0:
        x=(idade[y],sexo[y],tensao[y],colesterol[y],batimento[y],temdoenca[y])
        y=y+1
        N=N-1
        paciente.append(x)

     

   
         
  
     

def distporsexo():
    N=len(idade)
    masc=0
    fem=0
   

    for p in paciente:
        idadep,sexop,tensaop,colesterolp,batimentop,temdoencap=p 
        if temdoencap=="1":
            if sexop=="M":
                masc=masc+1

            if sexop=="F":
                fem=fem+1    

                
    print("O número de doentes do sexo masculino é {}".format(masc)) 
    print("O número de doentes do sexo feminino é {}".format(fem))
    

def distporidade():
    a=0
    b=0
    c=0
    d=0
    e=0
    f=0
    g=0
    h=0
    i=0
    j=0

    for p in paciente:
        idadep,sexop,tensaop,colesterolp,batimentop,temdoencap=p 
        if temdoencap=="1":
            if "30"<idadep<"34":
                a=a+1
            if "35"<idadep<"39":
                b=b+1    
            if "40"<idadep<"44":
                c=c+1    
            if "45"<idadep<"49":
                d=d+1  
            if "50"<idadep<"54":  
                e=e+1
            if "55"<idadep<"59":  
                f=f+1      
            if "60"<idadep<"64":
                g=g+1    
            if "65" <idadep< "69":
                h=h+1   
            if "70" <idadep< "74":
                i=i+1   
            if "75" <idadep< "79":
                j=j+1   
    


    
    
    
    print("O número de doentes entre os [30/34] anos é {}.".format(a))
    print("O número de doentes entre os [35/39] anos é {}.".format(b))
    print("O número de doentes entre os [40/44] anos é {}.".format(c))
    print("O número de doentes entre os [45/49] anos é {}.".format(d))
    print("O número de doentes entre os [50/54] anos é {}.".format(e))
    print("O número de doentes entre os [55/59] anos é {}.".format(f))
    print("O número de doentes entre os [60/64] anos é {}.".format(g))
    print("O número de doentes entre os [65/69] anos é {}.".format(h))
    print("O número de doentes entre os [70/74] anos é {}.".format(i))
    print("O número de doentes entre os [75/79] anos é {}.".format(j))



def nivelcol():
    nivel0=0
    indice=[]
    nivel=[]
    listanivel=[]
    

    for p in colesterol:
        if p == "0":
          nivel0=nivel0+1    
    
    q=60
    a=1
    n=100
    n1=110

    while q>0:
        x=(n,n1,a)
        nivel.append(x)
        a=a+1
        n=n+10
        n1=n1+10
        q=q-1

   
    
    
    for p in paciente:
       idadep,sexop,tensaop,colesterolp,batimentop,temdoencap=p 
       for o in nivel:
         min,max,ind=o
         if temdoencap=="1" and int(colesterolp)>=int(min) and int(colesterolp)<=int(max):
            listanivel.append(ind)

   


    listalen=[]
    for i in listanivel:
        if i not in listalen:
            listalen.append(i)

  


    g=len(listalen)    
    m=1
    print("Há {} doentes com indice 0 de colesterol".format(nivel0))
    while g>0:
       
        z=listanivel.count(m)
        m+=1
        print("Há {} doentes com indice {} de colesterol".format(z,m))
        g=g-1





def tabeladistsexo():
    
    dados={
        "Sexo":["Masculino","Feminino"],
        "Numero":[428,40]
    }

    df=pd.DataFrame(dados, columns=["Sexo","Numero"],)
    print(df)



print("Bem Vindo")

print("------------Menu--------------\n (1) Ler Lista \n (2) Distribuição por Sexo \n (3) Distribuição por idades \n (4) Nivel colesterol \n (5) Tabela \n (0) Sair")
ini=int(input("Escolhe a opção que pretendes"))
while ini!=0:
    if ini==1:
        lerlista()
        ini=int(input("Escolhe a opção que pretendes"))

    elif ini==2:
        distporsexo()  
        ini=int(input("Escolhe a opção que pretendes"))

    elif ini==3:
        distporidade()  
        ini=int(input("Escolhe a opção que pretendes"))   

    elif ini==4:
        nivelcol()  
        ini=int(input("Escolhe a opção que pretendes"))

    elif ini==5:
        tabeladistsexo()   
        ini=int(input("Escolhe a opção que pretendes"))   




if ini==0:
    print("Até à Próxima")



#Um programa que ao executar apresenta as tabelas pedidas é o Pandas

  
```
